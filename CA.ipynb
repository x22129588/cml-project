```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df = pd.read_csv('C:\\Users\\Hp\\Downloads\\healthcare-dataset-stroke-data.csv')
```

```python
df.head()
```

```python
#mapping evermarried yes to 1 and no to 0
my_mp = {'Yes': 1, 'No': 0}
df['ever_married'] = df['ever_married'].map(my_mp)
df.head(n=15)
```

```python
#mp_map = {'Female': 0, 'Male': 1}
df['gender'] = df['gender'].map({'Female': 0, 'Male': 1})
df.head(n=15)
```

```python
print (df.Residence_type.value_counts())
```

```python
# making the rural to 0 and urban to 1
df['Residence_type'] = df['Residence_type'].map({'Rural': 0, 'Urban': 1})
df.head(n=10)
```

```python
print (df.smoking_status.value_counts())
```

```python
print (df.work_type.value_counts())
```

```python
# assigning pprivatejob =0 self_employed=1 children=2 govt_job=3 never_worked=4
df['work_type'] = df['work_type'].map({'Private': 0, 'Self-employed': 1, 'children': 2, 'Govt_job': 3, 'Never_worked': 4})
print(df.head(n=10))
print (df.work_type.value_counts())
```

```python
print (df.smoking_status.value_counts())
## mapping never smoked to 0, smokes to 1, formely smokes 2, unknown to 3
df['smoking_status'] = df['smoking_status'].map({"never smoked": 0, 'smokes': 1, "formerly smoked": 2, 'Unknown': 3})
df.head(n=10)
print (df.smoking_status.value_counts())
print(df.head(n=10))
```
